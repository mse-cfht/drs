```python
# Import Requisite Python Packages
# Further python configuration and instructions will be written.
import vaex
import matplotlib
import matplotlib.pylab as plt
import numpy as np
```

```python
# Set-Up Configuration
%matplotlib inline
np.warnings.filterwarnings('ignore')
```

```python
# Open File
# File "gaia_esa_posb_query_fixed.hdf5" contains all Gaia EDR3 stars with G = 16-20 
# in the range b (Galactic Lat) = 20-90 degrees
# These eDR3 data represent only a subset of the full eDR3 data available for the selected stars.
# The file is 4.2 GB in size and contains 54,248,318 objects
# Additional quality control criteria will be employed for culling purposes. 
df = vaex.open("gaia_esa_posb_query_fixed.hdf5")
```

```python
#Cursory File Verification
len(df)
```

```python
#Visualization of the density of the sources on the sky by plotting a 2D histogram of RA vs Dec:
df.plot("ra", "dec", figsize=(9,6))
```

```python
#Plot of the log of the density
df.plot("ra", "dec", f="log1p", figsize=(9,6))
```

```python
#Plot of the sky density in galactic coordinates with the use of healpy
df.healpix_plot(figsize=(9,6), f="log1p")
```

```python
#Plot of the sky density in ecliptic coordinates with the use of healpy
df.healpix_plot(figsize=(9,6), f="log1p", healpix_output="ecliptic")
```

```python
#Initial attempt of an interactive plot with the employment of plot_widgets 
df.plot_widget("ra", "dec")
```
